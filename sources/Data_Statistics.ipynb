```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

pd.set_option('display.float_format', '{:.1f}'.format)
```

```python
df_country = pd.read_csv("../datasets/economy_asia_countries.csv", index_col=0)

df_country.head(3)
```

```python
df_country.describe()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(19, 8))
ax1.boxplot(x=df_country[["Exports (% of GDP)", "Imports (% of GDP)"]])
ax1.set_xticklabels(["Exports (% of GDP)", "Imports (% of GDP)"])
ax2.boxplot(x=df_country[["FDI (% of GDP)", "Unemployment (% of total labor force)"]])
ax2.set_xticklabels(["FDI (% of GDP)", "Unemployment (% of total labor force)"])
ax3.boxplot(x=df_country["Inflation (%)"])
ax3.set_xticklabels(["Inflation (%)"])
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(19, 12))
i = 0
j = 0
for col in df_country.describe().columns:
    sns.histplot(ax=axes[i, j],data=df_country[col])
    j = j + 1
    if(j == 3):
        i = i + 1
        j = 0

plt.show()
```
